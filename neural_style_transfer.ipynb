```python
import sys
sys.path.append('..')
from src.NST_Model import NST_Model
from src.VGG19 import VGG19
```

```python
# Load the VGG19 model
vgg19 = VGG19()

# Load the NST model
nst = NST_Model(vgg19)
```

```python
# Load the content image
content_image = nst.image_loader('data/content/cats.jpg')

# Load the style image
style_image = nst.image_loader('data/style/expressionism.jpg')
```

```python
# Show the content image
nst.imshow(content_image, title='Content Image')

# Show the style image
nst.imshow(style_image, title='Style Image')
```

```python
# Generate Random Noise Image
input_image = nst.generate_noise_image(content_image)

# Show the random noise image
nst.imshow(input_image, title='Random Noise Image')
```

```python
# Generate the stylized image
generated_image = nst.train(content_image, style_image, input_image, epochs=200)
```

```python
nst.save_and_display_image(generated_image, 'cats_expressionism.jpg')
```
